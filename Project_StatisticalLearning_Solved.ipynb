### Question 7

Lets define **Goal difference** as the difference between the **GoalsFor** and **GoalsAgainst**. And it can take negative values also.

Print all the goal differences for all the teams.

Write a function with name **goal_diff_count** which should return all the teams with their Goal Differences.

Write a function with name **max_goal_diff_team** which should return the team with max Goal Difference.

Write a function with name **min_goal_diff_team** which should return the team with min Goal Difference.



```python
laliga.insert(10, 'Goal Diff', ((laliga["GoalsFor"]).astype(int))-((laliga["GoalsAgainst"]).astype(int)))
```

```python
#max_goal_diff_team():
def max_goal_diff_team():
    maxGoalDiff=(abs(laliga["Goal Diff"])).max()
    laliga4=laliga.set_index("Goal Diff", drop=False)
    return laliga4.loc[maxGoalDiff].loc["Team"]

max_goal_diff_team()
```

```python
#min_goal_diff_team():
def min_goal_diff_team():
    minGoalDiff=(abs(laliga["Goal Diff"])).min()
    laliga4=laliga.set_index("Goal Diff", drop=False)
    return laliga4.loc[minGoalDiff].loc["Team"]

min_goal_diff_team()
```

```python
#goal_diff_count():
def goal_diff_count():
    "This function prints the goal differences of all teams"
    return (laliga.loc[:, ['Team','Goal Diff']])

goal_diff_count()
```
